```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([0, 0.15, 0.30])
y1 = np.array([5.17, 4.85, 4.61])
y2 = np.array([7.00, 6.52, 6.44])
```

```python
sns.set_style("whitegrid")
sns.color_palette("rocket")
plt.rcParams['font.sans-serif'] = ['SimHei']

fig, ax = plt.subplots(figsize=(6, 4), dpi=400)
sns.lineplot(x=x, y=y1, ax=ax, linewidth=2, marker='o', markersize=5, label="7 days")
sns.lineplot(x=x, y=y2, ax=ax, linewidth=2, marker='o', markersize=5, label="28 days")

for i in range(len(x)):
    plt.annotate(y1[i], xy=(x[i], y1[i]), xytext=(x[i], y1[i]+0.1))
    plt.annotate(y2[i], xy=(x[i], y2[i]), xytext=(x[i], y2[i]+0.1))

ax.set_xlabel("粉煤灰掺量", fontsize=8)
ax.set_ylabel("抗拉强度(MPa)", fontsize=8)
ax.tick_params(labelsize=8)
plt.show()
```
