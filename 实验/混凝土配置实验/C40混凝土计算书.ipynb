## 配置强度计算

混凝土设计强度等级小于C60，配置强度应按照下述的强度约束公式确定：
$$
f_{cu, 0} \ge f_{cu, k} + 1.645 \sigma
$$
其中$f_{cu, 0}$为混凝土配置强度，$f_{cu, k}$为混凝土立方体抗压强度标准值，$\sigma$为混凝土强度标准差，单位均为MPa。

```python
f_cuk = 40.0   # 混凝土强度标准值 # MPa
sigma = 5.0    # 混凝土强度标准差 # MPa
f_ce0 = 60.0   # 基准水泥强度     # MPa
rho_c = 3150.0 # 水泥表观密度     # kg/m^3
```

## 配合比设计

### 掺合料相关系数

```python
# Fly Ash         # 粉煤灰
ratio_fa = 0.18   # 掺量
gamma_fa = 0.4    # 胶凝系数
rho_fa   = 2350.0 # 表观密度，kg/m^3

# Mineral Powder  # 矿渣粉
ratio_mp = 0.19   # 掺量
gamma_mp = 0.9    # 胶凝系数
rho_mp   = 2950.0 # 表观密度，kg/m^3

# Silica Fume     # 硅灰
ratio_sf = 0.03   # 掺量
gamma_sf = 1.05   # 胶凝系数
rho_sf   = 2300.0 # 表观密度，kg/m^3
```

```python
V_air          = 0.015  # m^3    # 单位空气体积
V_stone_heap   = 0.59   # m^3    # 松堆石子体积分数

rho_stone_heap = 1550.0 # kg/m^3 # 石子堆积密度
rho_stone      = 2800.0 # kg/m^3 # 石子表观密度
rho_sand       = 2650.0 # kg/m^3 # 砂的表观密度
rho_water      = 1000.0 # kg/m^3 # 水的表观密度

# Sand/Mortar Ratio
ratio_sm = 0.42 # 砂浆中砂的占比
```

```python
M_stone  = V_stone_heap * rho_stone_heap # kg  # 单位石子质量
V_stone  = M_stone / rho_stone           # m^3 # 单位石子体积
V_mortar = 1 - V_stone                   # m^3 # 单位砂浆体积
V_sand   = V_mortar * ratio_sm           # m^3 # 单位砂的体积
M_sand   = V_sand * rho_sand             # kg  # 单位砂的质量
V_paste  = V_mortar - V_sand             # m^3 # 单位浆体体积
```

```python
# 计算得出的单位石子体积应等于设定的单位粗骨料体积
print(V_stone)            # m^3 # 单位石子体积
V_coarse_aggregate = 0.32 # m^3 # 单位粗骨料体积
```

## 配合计算

### 混凝土配置强度计算

```python
f_cu0_min = f_cuk + 1.645 * sigma

print("混凝土配置强度最小值: %.4f MPa" % f_cu0_min)
```

```python
f_cu0 = 55.0

print("取混凝土配置强度为: %.4f MPa" % f_cu0)
```

### 水胶比计算

```python
ratio_wb = 0.42 * f_ce0 \
            * (1 - ratio_fa - ratio_mp - ratio_sf + ratio_fa * gamma_fa + ratio_mp * gamma_mp + ratio_sf * gamma_sf) \
            / (f_cu0 + 1.2)

print("水胶比: %.4f %%" % (ratio_wb * 100))
```

### 胶凝材料表观密度计算

```python
rho_binder = 1 / (ratio_fa / rho_fa + ratio_mp / rho_mp + ratio_sf / rho_sf + (1 - ratio_fa - ratio_mp - ratio_sf) / rho_c)

print("胶凝材料表观密度: %.4f kg/m^3" % rho_binder)
```

### 胶凝材料质量计算

```python
M_binder = (V_paste - V_air) / (1 / rho_binder + ratio_wb / rho_water)

print("胶凝材料质量: %.4f kg" % M_binder)
```

### 水的质量计算

```python
M_water = M_binder * ratio_wb

print("水的质量: %.4f kg" % M_water)
```

### 掺合料的质量计算

```python
M_fa, M_mp, M_sf = ratio_fa * M_binder, ratio_mp * M_binder, ratio_sf * M_binder

print("粉煤灰质量: %.4f kg" % M_fa)
print("矿渣粉质量: %.4f kg" % M_mp)
print("硅灰的质量: %.4f kg" % M_sf)
```

### 水泥质量计算

```python
M_c = M_binder - M_fa - M_mp - M_sf
print("水泥质量: %.4f kg" % M_c)
```

### 浆骨比

```python
ratio_mb = (M_water + M_fa + M_mp + M_sf + M_c) / (M_stone + M_sand)
print("浆骨比: %.4f" % ratio_mb)
```
