```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from pickle import load
```

```python
x = list(range(1, 201))
gat_loss_hist = load(open('./logs/gat/loss_hist.pkl', 'rb'))
gcn_loss_hist = load(open('./logs/gcn/loss_hist.pkl', 'rb'))
node2vec_loss_hist = load(open('./logs/node2vec/loss_hist.pkl', 'rb'))
gin_loss_hist = load(open('./logs/gin/loss_hist.pkl', 'rb'))
df = pd.DataFrame({'X': x, 'GAT': gat_loss_hist, 'GCN': gcn_loss_hist, 'node2vec': node2vec_loss_hist, 'GIN': gin_loss_hist}).set_index('X')


sns.set_style("darkgrid")
sns.set_palette("flare")
fig, ax = plt.subplots(figsize=(8, 4), dpi=400)
sns.lineplot(data=df, ax=ax, palette="CMRmap", dashes=False)

ax.set_xlabel("Epoch", fontsize=8)
ax.set_ylabel("Loss", fontsize=8)
ax.tick_params(labelsize=8)
ax.set_title("Loss Curve", fontsize=12)

plt.show()
```
